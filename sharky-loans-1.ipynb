```python
# loans (of the sr manzano collection) expiring in the next 36 hours
import pandas as pd
import datetime
import timedelta
```

```python
df = pd.read_csv('data/nft-sr-m.csv', header=0, index_col=False).astype('object')
```

```python
span_start = datetime.datetime.now() - timedelta.Timedelta(days=7)
span_end = datetime.datetime.now() - timedelta.Timedelta(days=7, hours=-36)
span_start_unix = int(span_start.timestamp())
span_end_unix = int(span_end.timestamp())
df_slice = df.query(f'op_3.isnull() and op_2!="Rescind" and op_2.notnull() and b_time_2>={span_start_unix} and b_time_2<={span_end_unix}').copy().sort_values(['b_time_2'], ascending=[False]).reset_index(drop=True)
df_slice.index += 1
```

```python
df_slice
```

```python
# borrower + hours_left
df_slice.loc[:, ['borrower_readable']]  = df_slice['borrower'].str[0:3] + '___' + df_slice['borrower'].str[-3:]
df_slice.loc[:, ['hours_left']] = round((df_slice['b_time_2'].astype(int) - span_start_unix)/60/60).astype(int)
```
